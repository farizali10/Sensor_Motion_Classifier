```python
input_file_path = r"/config/workspace/data/sensor_placement/bending2/dataset4.csv"
output_file_path = r'/config/workspace/data/sensor_placement/bending2/new_dataset4.csv'
```

```python
with open(input_file_path,'r') as infile:
    lines = infile.readlines()

    csv = []
    for i in range(len(lines)):
        csv.append(lines[i].replace(" ", ","))
```

```python
# Remove '\n' from each string in the list
new_csv = [string.replace(',\n', '') for string in csv]
```

```python
a = []
for row in new_csv:
    a.append(row.split(","))
```

```python
import csv

# Open the CSV file in write mode and create a CSV writer
with open(output_file_path, 'w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)

    # Write each row of data to the CSV file
    for row in a:
        csv_writer.writerow(row)
```
