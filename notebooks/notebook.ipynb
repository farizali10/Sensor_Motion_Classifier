```python
import pandas as pd
import os

def csv_reader(org_path, custom_df_name, num_columns, folder_path):
    """
    Arguments:
        org_path(str): Orginal dataset path,
        custom_df_name(str): New dataframe name,
        num_columns(int): Number of columns in desired dataset,
        folder_path(str): New dataset's path to be saved.
        
    Import Requirements:
        - pandas
        - os
    """
    
    try:
        # Check if the target folder exists, and create it if it doesn't
        if not os.path.exists(folder_path):
            os.makedirs(folder_path)
            
        with open(org_path, "r") as file:
            lines = file.readlines()

        delimiter = ','  # Replace with the appropriate delimiter
        rows = [line.strip().split(delimiter) for line in lines]

        # Define the number of columns in your DataFrame
        total_columns = num_columns  # Replace with the appropriate number

        # Initialize an empty DataFrame with the specified number of columns
        custom_df = pd.DataFrame(columns=[f"Column{i+1}" for i in range(num_columns)])

        # Iterate through the rows and add them to the DataFrame
        for row in rows:
            # Pad or truncate the row to match the number of columns
            row = row[:num_columns] if len(row) >= num_columns else row + [''] * (num_columns - len(row))
            custom_df = custom_df.append(pd.Series(row, index=custom_df.columns), ignore_index=True)
            
        # Save the DataFrame to a CSV file
        csv_filename = f"{custom_df_name}.csv"
        custom_df.to_csv(os.path.join(folder_path, csv_filename), index=False)

    except Exception as e:
        print(e)
```

```python
import pandas as pd
import os

def csv_reader(org_path, custom_df_name, num_columns, output_folder):
    """
    Arguments:
        org_path(str): Original dataset path,
        custom_df_name(str): New dataframe name,
        num_columns(int): Number of columns in the desired dataset,
        output_folder(str): Output directory for saving the files.
        
    Import Requirements:
        - pandas
        - os
    """
    
    try:
        with open(org_path, "r") as file:
            lines = file.readlines()

        delimiter = ','  # Replace with the appropriate delimiter
        rows = [line.strip().split(delimiter) for line in lines]

        # Define the number of columns in your DataFrame
        total_columns = num_columns  # Replace with the appropriate number

        # Initialize an empty DataFrame with the specified number of columns
        custom_df = pd.DataFrame(columns=[f"Column{i+1}" for i in range(num_columns)])

        # Iterate through the rows and add them to the DataFrame
        for row in rows:
            # Pad or truncate the row to match the number of columns
            row = row[:num_columns] if len(row) >= num_columns else row + [''] * (num_columns - len(row))
            custom_df = custom_df.append(pd.Series(row, index=custom_df.columns), ignore_index=True)
            
        # Save the DataFrame to a CSV file in the output folder
        csv_filename = f"{custom_df_name}.csv"
        custom_df.to_csv(os.path.join(output_folder, csv_filename), index=False)

    except Exception as e:
        print(e)

def func_file_list(folder_path):
    """
    Arguments:
        folder_path (str): Folder's content to be read.
    Import Requirements:
        - os
    """
    try:
        path_list = []

        # List all files in the folder
        contents = os.listdir(folder_path)

        # Print the contents
        for item in contents:
            path = os.path.join(folder_path, item)
            path_list.append(path)
            
        return path_list
    
    except Exception as e:
        print(e)

# Set the parent folder path
parent_folder = "D:\\Data-Science-D-drive\\Datasets-D-drive\\sensor_placement"

# Iterate through folders in the parent folder
for folders in func_file_list(parent_folder):
    try:
        print("- " + os.path.basename(folders))  # Get the folder name without the full path
        print("="*100)

        # Create an 'output' folder within the parent folder
        output_folder = os.path.join(folders, "output")
        os.makedirs(output_folder, exist_ok=True)
        
        for files in func_file_list(folders):
            print(files)
            
            # Extract the file name without the full path
            file_name = os.path.basename(files)
        
            csv_reader(files, "new_" + file_name, 7, output_folder)
        
        print("="*100)
    except Exception as e:
        print(e)
```
